```python
import numpy

data = numpy.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
print(data)

import matplotlib.pyplot
image = matplotlib.pyplot.imshow(data)
matplotlib.pyplot.show()
```

```python
# average inflammation over time
ave_inflammation = numpy.mean(data, axis=0) # axis 0 refers to the rows
ave_plot = matplotlib.pyplot.plot(ave_inflammation)
matplotlib.pyplot.show()
```

```python
# The maximum inflammation value rises and falls linearly
max_plot = matplotlib.pyplot.plot(numpy.amax(data, axis=0))
matplotlib.pyplot.show()
```

```python
# The minimum inflammation seems to be a step function
min_plot = matplotlib.pyplot.plot(numpy.amin(data, axis=0))
matplotlib.pyplot.show()
```

```python
# Grouping plots
import numpy
import matplotlib.pyplot

data = numpy.loadtxt(fname='data/inflammation-01.csv', delimiter=',')
fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0)) # creates a space for the figure
# people often 'import matplotlib.pyplot as plt'
# so it would then be plt.figure() instead 

# each figure is added to 'fig' object with fig.add_subplot method 
# add_subplot takes 3 arguments: 
# the first argument is how many rows of subplots there are
# the second argument is how many columns 
# the third argument is which subplot your variable references, left to right
axes1 = fig.add_subplot(1, 3, 1)
axes2 = fig.add_subplot(1, 3, 2)
axes3 = fig.add_subplot(1, 3, 3)

# axes can be titled using set_ylabel and set_xlabel 
axes1.set_ylabel('average')
axes1.set_xlabel('time (days)')
axes1.plot(numpy.mean(data, axis=0))

axes2.set_ylabel('max')
axes2.plot(numpy.amax(data, axis=0))

#axes3.set_ylabel('min')
#axes3.plot(numpy.amin(data, axis=0))
# original code above, new code below shows how to manually set y axis scale
min_data = numpy.amin(data, axis=0)
axes3.set_ylabel('min')
axes3.plot(min_data)
axes3.set_ylim(numpy.amin(min_data), numpy.amax(min_data) * 2.1)

# if the below is left out, it will actually squeeze them in more closely
fig.tight_layout()

matplotlib.pyplot.savefig('inflammation.png')
matplotlib.pyplot.show()
```

```python
# An automated approach to scaling the y-axis
min_data = numpy.amin(data, axis=0)
axes3.set_ylabel('min')
axes3.plot(min_data)
axes3.set_ylim(numpy.amin(min_data), numpy.amax(min_data) * 1.1)
```

```python
# Changing the draw style 
# Change the drawing style so matplotlib doesn't draw a straight line between points

import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt(fname='data/inflammation-01.csv', delimiter=',')

fig = plt.figure(figsize=(6.0, 6.0))

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

axes1.set_ylabel('average')
axes1.plot(np.mean(data, axis=0), drawstyle='steps-mid')

axes2.set_ylabel('max')
axes2.plot(np.amax(data, axis=0), drawstyle='steps-mid')

axes3.set_ylabel('min')
axes3.plot(np.amin(data, axis=0), drawstyle='steps-mid')

# add another chart for standard deviation 
axes4.set_ylabel('std deviation')
axes4.plot(np.std(data,axis=0), drawstyle='steps-mid')

fig.tight_layout()

plt.show()
```
